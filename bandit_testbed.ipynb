```python
from bandit_testbed import KArmedBanditTestbed
from bandit_algorithms import GreedyBandit, EpsilonGreedyBandit
import scipy
```

```python
algos = {
    'greedy': GreedyBandit(k_arms=10),
    'epsilon_0.1': EpsilonGreedyBandit(k_arms=10, epsilon=0.1),
}

testbed = KArmedBanditTestbed(k_arms=10,
                             time_steps=1000,
                             episodes=2000,
                             algorithms=algos)
```

```python
results = testbed.run()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1)

learning_curves = {
    'greedy': pd.DataFrame(results['greedy']).mean(),
    'epsilon_0.1': pd.DataFrame(results['epsilon_0.1']).mean(),
}

pd.DataFrame(learning_curves).plot(figsize=(10, 5), ax=ax, lw=0.5)

ax.set_ylim([0, 1.5])
ax.set_yticks([0, 0.5, 1, 1.5])
ax.set_xticks([1, 250, 500, 750, 1000])
```
